```python
#------------------
# import packages
#------------------

import requests
import pycountry
import pandas as pd
import time  # for sleep/backoff

#-------------------
# setup parameters
#-------------------

# 1) list of country codes (ISO alpha-2)
country_codes = [country.alpha_2 for country in pycountry.countries]

# 2) base API URL for IMF data
base_url = 'http://dataservices.imf.org/REST/SDMX_JSON.svc/'

# 3) dataset code
dataset_code = 'IFS'

# 4) indicator codes for exchange rates and GDP 
#    - We still use EDNA_USD_XDC_RATE for exchange rates
#    - For GDP:
#         Annual = NGDP_XDC
#         Quarterly = NGDP_NSA_XDC
exchange_rate_indicator = 'EDNA_USD_XDC_RATE'
annual_gdp_indicator = 'NGDP_XDC'
quarterly_gdp_indicator = 'NGDP_NSA_XDC'

# 5) indicator labels
indicator_labels = {
    'exchange': 'Exchange Rate (US Dollar per Domestic Currency)',
    'gdp': 'GDP (Nominal, Unadjusted)'
}

# 6) frequencies: annual (A) and quarterly (Q)
frequencies = {
    'A': 'Annual',
    'Q': 'Quarterly'
}

#--------------------------------
# function to retrieve IMF data
#--------------------------------

def get_imf_data(reporter,
                 dataset=dataset_code,
                 indicator='EDNA_USD_XDC_RATE',
                 freq='A',
                 startPeriod='1980',
                 endPeriod='2024',
                 max_retries=5,
                 backoff_factor=3):
    """
    Retrieve IMF data for a given reporter country, with basic retry logic.
    
    Parameters:
    -----------
    reporter : str
        ISO alpha-2 code for the reporter country (e.g., 'GB', 'US', 'CN', etc.)
    dataset : str
        IMF dataset code (e.g., 'IFS')
    indicator : str
        IMF indicator code (e.g., 'EDNA_USD_XDC_RATE' for exchange)
    freq : str
        Frequency ('A' for annual, 'Q' for quarterly, 'M' for monthly)
    startPeriod : str
        Starting period (e.g., '1980')
    endPeriod : str
        Ending period (e.g., '2024')
    max_retries : int
        Maximum number of retry attempts for transient errors
    backoff_factor : int or float
        Multiplier for the backoff (sleep time) between retries
    
    Returns:
    --------
    data : dict or list or None
        Parsed JSON series data, or None if we can't retrieve valid data.
    """
    # construct the SDMX key: 'CompactData/<dataset>/<freq>.<reporter>.<indicator>'
    key = f'CompactData/{dataset}/{freq}.{reporter}.{indicator}'
    full_url = f'{base_url}{key}?startPeriod={startPeriod}&endPeriod={endPeriod}'
    
    for attempt in range(max_retries):
        print(f"[Attempt {attempt+1}] Fetching data for {reporter} at {freq} freq: {full_url}")
        try:
            response = requests.get(full_url, headers={"Accept": "application/json"}, timeout=30)
            response.raise_for_status()  # Raises HTTPError if not 200
            json_data = response.json()
            
            # Navigate to the series element
            data = json_data['CompactData']['DataSet']['Series']
            return data

        except requests.exceptions.RequestException as e:
            # network issues, timeouts, or HTTP errors
            if attempt == max_retries - 1:
                print(f"Final network error (no more retries) for {reporter}: {e}")
                return None
            else:
                wait_seconds = backoff_factor * (attempt + 1)
                print(f"Network error for {reporter}, retrying in {wait_seconds} seconds: {e}")
                time.sleep(wait_seconds)

        except KeyError as e:
            # JSON structure missing the expected keys (e.g., 'Series' not found)
            print(f"JSON structure error (missing keys) for {reporter}: {e}")
            return None

        except ValueError as e:
            # JSON parse error (possibly rate limit returning non-JSON response)
            if attempt == max_retries - 1:
                print(f"Final JSON parse error for {reporter}: {e}")
                return None
            else:
                wait_seconds = backoff_factor * (attempt + 1)
                print(f"JSON parse error for {reporter}, retrying in {wait_seconds} seconds: {e}")
                time.sleep(wait_seconds)
    
    # if it exhausts the loop without returning, return None
    return None

#--------------------------
# retrieve and store data
#--------------------------

data_dicts = {}  # will hold results for each (indicator, frequency) combination

for freq_code, freq_label in frequencies.items():
    
    # 1) exchange rates
    exch_data_dict = {}
    print(f"\nFetching {freq_label} Exchange Rate data...")
    for rep in country_codes:
        data = get_imf_data(
            reporter=rep,
            dataset=dataset_code,
            indicator=exchange_rate_indicator,  # always EDNA_USD_XDC_RATE
            freq=freq_code,
            startPeriod='1947',
            endPeriod='2024'
        )
        if data:
            exch_data_dict[rep] = data
        else:
            print(f"No exchange rate data returned for {rep} at {freq_label} frequency.")
    data_dicts[f"exchange_{freq_label}"] = exch_data_dict

    # 2) GDP
    gdp_data_dict = {}
    print(f"\nFetching {freq_label} GDP data...")
    
    # choose the GDP indicator based on frequency
    if freq_code == 'A':
        # annual
        chosen_gdp_indicator = annual_gdp_indicator
    else:
        # quarterly
        chosen_gdp_indicator = quarterly_gdp_indicator
    
    for rep in country_codes:
        data = get_imf_data(
            reporter=rep,
            dataset=dataset_code,
            indicator=chosen_gdp_indicator,  # NGDP_XDC for annual, NGDP_NSA_XDC for quarterly
            freq=freq_code,
            startPeriod='1947',
            endPeriod='2024'
        )
        if data:
            gdp_data_dict[rep] = data
        else:
            print(f"No GDP data returned for {rep} at {freq_label} frequency.")
    data_dicts[f"gdp_{freq_label}"] = gdp_data_dict

#----------------------
# process series data
#----------------------

def process_series_data(data_dict, indicator_label, freq_label):
    """
    Convert the IMF series data into a pandas DataFrame.
    
    Parameters:
        data_dict (dict): Dictionary keyed by reporter country codes.
        indicator_label (str): Human-readable indicator labe
.
        freq_label (str): 'Annual' or 'Quarterly'
        
    Returns:
        df (pd.DataFrame): DataFrame with columns [reporter, indicator, frequency, time_period, value]
    """
    records = []
    for rep, series in data_dict.items():
        # The returned series might be a list or a dict
        if isinstance(series, list):
            for s in series:
                obs = s.get('Obs')
                if isinstance(obs, list):
                    for o in obs:
                        records.append({
                            'reporter': rep,
                            'indicator': indicator_label,
                            'frequency': freq_label,
                            'time_period': o.get('@TIME_PERIOD'),
                            'value': o.get('@OBS_VALUE')
                        })
                elif isinstance(obs, dict):
                    records.append({
                        'reporter': rep,
                        'indicator': indicator_label,
                        'frequency': freq_label,
                        'time_period': obs.get('@TIME_PERIOD'),
                        'value': obs.get('@OBS_VALUE')
                    })
        elif isinstance(series, dict):
            obs = series.get('Obs')
            if isinstance(obs, list):
                for o in obs:
                    records.append({
                        'reporter': rep,
                        'indicator': indicator_label,
                        'frequency': freq_label,
                        'time_period': o.get('@TIME_PERIOD'),
                        'value': o.get('@OBS_VALUE')
                    })
            elif isinstance(obs, dict):
                records.append({
                    'reporter': rep,
                    'indicator': indicator_label,
                    'frequency': freq_label,
                    'time_period': obs.get('@TIME_PERIOD'),
                    'value': obs.get('@OBS_VALUE')
                })
    df = pd.DataFrame(records)
    if 'time_period' in df.columns:
        df['time_period'] = pd.to_datetime(df['time_period'], errors='coerce')
    return df

#-----------------------
# save the data to CSV
#-----------------------

for key, data_dict in data_dicts.items():
    # example key: "exchange_annual" or "gdp_quarterly"
    indicator_key, freq_label = key.split('_', maxsplit=1)
    indicator_label = indicator_labels.get(indicator_key, indicator_key.capitalize())
    
    df = process_series_data(data_dict, indicator_label, freq_label)
    
    csv_filename = f"../../data/raw/imf_{indicator_key}_{freq_label.lower()}.csv"
    df.to_csv(csv_filename, index=False)
    print(f"Saved {indicator_label} ({freq_label}) data to {csv_filename}")
```
